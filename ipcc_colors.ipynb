# Colormaps from IPCC AR6

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# get a list of the text files with the RGB data in the data directory
!ls ./ipcc_colors/*.txt
```

```python
# read the selected txt file - text files from: https://github.com/IPCC-WG1/colormaps
rgb_data_in_the_txt_file = np.loadtxt("./ipcc_colors/slev_div.txt")

# create the colormap
my_colormap = mcolors.LinearSegmentedColormap.from_list('colormap', rgb_data_in_the_txt_file)
```

Test the colormap:

```python
N = 11
x = np.linspace(0., 10., N)
y = np.linspace(0., 10., N)
z = np.random.uniform(-2., 2., size=(N,N))

levels = 31
plt.contourf(x, y, z, levels, cmap=my_colormap) # colormap adapts to the number of levels in levels
plt.colorbar()
plt.show()
```
